# Preprocessing data

With the ultimate goal of clustering and classifying postcodes based on their economic and financial situation, this notebook was designed to preprocess and clean data sourced from official entities, namely ABS (Australian Bureau of Statistics) and ATO (Australian Taxation Office). Both datasets contain aggregated information presented in average and median formats.

Key Features:

* **Data Loading**: The initial section outlines the process of importing the aggregated financial and economic datasets pertaining to Australian postcodes, setting the stage for the subsequent steps.

* **Data Cleaning**: This segment addresses potential inconsistencies, anomalies, missing values, and human errors. Any unusual values or outliers will be replaced by NaN and then imputed, only if they constitute a non-significant portion (< 10%). For outliers, we have to consider that we are dealing with financial data, so right skew distributions are expected.

* **Feature Engineering:** This section is dedicated to enhancing the depth of the dataset by creating or adjusting features that better encapsulate the underlying patterns.


Data scaling will be done in the model notebook, since two scaling methods will be tried.

```python
#Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Functions

```python
import matplotlib.pyplot as plt

def plot_hist_box(dataframe, column, bins):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw={'width_ratios': [3, 1]})
    
    # Plot histogram on the left
    ax[0].hist(dataframe[column].dropna(), bins=bins, color='#86bf91', rwidth=0.8)
    ax[0].set_title(f'Histogram of {column}', fontsize=12)
    ax[0].set_xlabel(column, fontsize=10)
    ax[0].set_ylabel('Frequency', fontsize=10)
    ax[0].grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.6)
    
    # Plot boxplot on the right
    ax[1].boxplot(dataframe[column].dropna(), vert=True, patch_artist=True, boxprops=dict(facecolor='#86bf91'))
    ax[1].set_title(f'Boxplot of {column}', fontsize=12)
    ax[1].set_xlabel(column, fontsize=10)
    ax[1].set_yticklabels([])  # Hide y-axis labels for boxplot to avoid duplication
    
    plt.tight_layout()
    plt.show()
```

# ABS data

## Load Data

```python
# Loading data

abs_data = pd.read_csv('../data/ABS_data_2021.csv', sep = ";", decimal = ".")
abs_data['postcode'].replace(r'[a-zA-Z%]', '', regex=True, inplace=True)
abs_data['postcode'] = abs_data['postcode'].astype(int)
print(abs_data.shape)
abs_data.head()
```

```python
abs_data.describe()
```

```python
abs_data.dtypes
```

```python
abs_data.isnull().sum()
```

To avoid losing information associated with postcodes, null values will be imputed using a spatial algorithm. This approach ensures that data integrity is maintained by leveraging the geographic relationships and patterns inherent in spatial data.

```python
abs_data.columns
```

## age

```python
plot_hist_box(abs_data, 'median_age', bins=20)
```

```python
abs_data['median_age'] = np.where((abs_data['median_age'] >= 25) & 
                                  (abs_data['median_age'] <= 65), 
                                  abs_data['median_age'], np.nan)


plot_hist_box(abs_data, 'median_age', bins=20)
```

# Final data

## Merge dataframes

```python
data_merge = pd.merge(left = abs_data, right = ato_data, on = 'postcode', how = 'outer')
print(data_merge.shape)
data_merge.head()
```

## Impute Null values

```python
#% of NaNs for each column
round(data_merge.isnull().sum()/data_merge.shape[0], 2)
```

**Given that consecutive postcodes typically represent geographically adjacent areas, it's reasonable to assume that they have similar socio-economic profiles. Therefore, we will adopt the 'ffill' method to address NaN values.**

```python
# Rows with more NaNs
data_merge = data_merge.fillna(method='ffill')

print(round(data_merge.isnull().sum()/data_merge.shape[0], 2))
print(data_merge.shape)
```

# Add city

```python
poa_data = pd.read_csv('../data/au_postcodes.csv', sep = ";")
print(poa_data.shape)
poa_data.head()
```

```python
final_data_region = pd.merge(left = poa_data, right = data_merge, on = 'postcode', how = 'inner')
final_data_region.head()
```

# Save data

```python
final_data_region.reset_index(drop=True).to_csv('../data/cleaned_data.csv', index=False)
```
